```python
import numpy as np
import matplotlib.pyplot as plt

# Given data points and their labels
data = np.array([[2, -1], [-1, 1], [2, 0.5], [0.2, -0.2], [0.5, -1], [2, 1]])
labels = np.array([1, 0, 1, 0, 1, 0])


# Function to plot the data and the hyperplane
def plot_hyperplane(w, b, data, labels, eta, iteration):
    # Plot the data points
    plt.scatter(data[:, 0], data[:, 1], c=labels, cmap=plt.cm.Paired)

    # Define the hyperplane
    x2 = np.linspace(-2, 3)
    x1 = -(b + w[1] * x2) / w[0]

    # Plot the hyperplane
    plt.plot(x1, x2, label=f"η={eta}, Iteration={iteration}")
    plt.xlim(-2, 3)
    plt.ylim(-2, 2)
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.title("Perceptron Hyperplane")
    plt.axhline(0, color="black", linewidth=0.5)
    plt.axvline(0, color="black", linewidth=0.5)
    plt.grid(True)
    plt.legend()
    plt.show()


# Perceptron training function
def perceptron_train(data, labels, eta, n_iter=10):
    w = np.array([1.0, 0.0])  # Initial weights
    b = 0.0  # Initial bias

    for iteration in range(n_iter):
        for i in range(len(data)):
            x = data[i]
            y = labels[i]
            y_hat = 1 if np.dot(w, x) + b > 0 else 0
            if y != y_hat:
                # Update rules for weights and bias
                w = w + eta * (y - y_hat) * x
                b = b + eta * (y - y_hat)
        # Plot the hyperplane at the current iteration
        plot_hyperplane(w, b, data, labels, eta, iteration + 1)

    return w, b


# Train the perceptron with learning rate eta = 2
w, b = perceptron_train(data, labels, eta=2, n_iter=1)

# Train the perceptron with learning rate eta = 0.5
w, b = perceptron_train(data, labels, eta=0.5, n_iter=1)
```

# Perceptron Algorithm

The perceptron is one of the simplest types of artificial neural networks and can be used for binary classification. It works by taking an input vector, applying a set of weights, adding a bias, and then passing the result through an activation function to produce an output.

## How it Works

The output of the perceptron is calculated as follows:

$$
y = f(\mathbf{w} \cdot \mathbf{x} + b)
$$

Where:
- $f$ is the activation function (often a step function).
- $\mathbf{w}$ represents the weight vector.
- $\mathbf{x}$ represents the input vector.
- $b$ is the bias term.

## Training the Perceptron

The perceptron is trained using the gradient descent method to update its weights and bias. The update rules are:

$$
\mathbf{w}_{\text{new}} = \mathbf{w}_{\text{old}} + \eta (y - \hat{y}) \mathbf{x}
$$
$$
b_{\text{new}} = b_{\text{old}} + \eta (y - \hat{y})
$$

Where:
- $\eta$ is the learning rate.
- $y$ is the true label.
- $\hat{y}$ is the predicted label.

## The Algorithm

1. Initialize weights $\mathbf{w}$ and bias $b$ to small random values.
2. For each training sample $(\mathbf{x}, y)$, perform the following steps:
   - Compute the predicted output $\hat{y} = f(\mathbf{w} \cdot \mathbf{x} + b)$.
   - Update the weights and bias if $\hat{y} \neq y$.
3. Repeat the above steps for a number of iterations or until the model converges.

## The Hyperplane

The perceptron defines a separating hyperplane with the equation:

$$
\mathbf{w} \cdot \mathbf{x} + b = 0
$$

This hyperplane divides the input space into two parts, each corresponding to one of the two possible classes.

## Learning Rates

Different learning rates can affect the training process:

- A higher learning rate (e.g., $\eta = 2$) may cause the model to converge quickly but can overshoot the minimum loss.
- A lower learning rate (e.g., $\eta = 0.5$) may result in more precise convergence but takes longer.

Adjusting the learning rate is crucial for the perceptron to learn effectively.
